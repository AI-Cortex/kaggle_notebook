```python
from pathlib import Path
import sys

path = '/kaggle/input/binary-classification/src'
if path not in sys.path:
    sys.path.append(path)
        
from data_generator import moons_data
import utils
import TF_model
import Torch_model

cfg_path = Path('/kaggle/input/binary-classification/config/config.yml')
cfg = utils.load_config(cfg_path)
```

# create Data


```python
(
    X_train_moons, 
    X_val_moons,
    X_test_moons, 
    y_train_moons, 
    y_val_moons, 
    y_test_moons
) = moons_data(cfg)

print(X_train_moons.shape, y_train_moons.shape)
print(X_val_moons.shape, y_val_moons.shape)
print(X_test_moons.shape, y_test_moons.shape)
```

# TF model


```python
tf_model = TF_model.ModelTF(cfg, 2, None)
tf_model.fit(x=X_train_moons, 
             y=y_train_moons,
             batch_size = cfg['batch_size'],
             validation_data=(X_val_moons, y_val_moons),
             epochs=cfg['epochs']
             )
```

```python
tf_result = tf_model.my_evaluate(X_test_moons, y_test_moons)
```

# Torch model


```python
torch_model = Torch_model.TorchWrapper(cfg['batch_size'])
torch_model.fit(X_train_moons, y_train_moons, X_val_moons, y_val_moons, cfg['epochs'])
```

```python
torch_result = torch_model.my_evaluate(X_test_moons, y_test_moons)
```

## TF Vs Torch 

```python
tf_result['accuracy'], torch_result['accuracy']
```

```python
print('TF classification report')
print(tf_result['classification_report'])
print('--------------------------------------------------')
print('Torch classification report')
print(torch_result['classification_report'])
```

```python
print('TF confusion matrix')
print(tf_result['confusion_matrix'])
print('--------------------------------------------------')
print('Torch confusion matrix')
print(torch_result['confusion_matrix'])
```
